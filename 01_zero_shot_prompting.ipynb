```python
from transformers import pipeline

# Load a zero-shot classification model
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")

print("Model loaded successfully!")
```

```python
# ── ZERO-SHOT PROMPTING ──────────────────────────────────────────
# Zero-shot means we give the model NO examples.
# We just describe the categories and let it figure it out.

text = "The patient has been experiencing chest pain and shortness of breath."

candidate_labels = ["medical", "sports", "technology", "politics"]

result = classifier(text, candidate_labels)

print("Text:", text)
print("\nResults:")
for label, score in zip(result['labels'], result['scores']):
    print(f"  {label}: {score:.2%}")
```

```python
# Try it on more examples to see how well it works

examples = [
    "The stock market crashed 500 points today.",
    "Messi scored a hat-trick in the Champions League.",
    "The new iPhone features a 48MP camera system.",
    "The government passed a new climate change bill."
]

labels = ["finance", "sports", "technology", "politics"]

print("ZERO-SHOT CLASSIFICATION RESULTS")
print("="*50)
for text in examples:
    result = classifier(text, labels)
    top_label = result['labels'][0]
    top_score = result['scores'][0]
    print(f"\nText: {text}")
    print(f"Predicted: {top_label} ({top_score:.2%} confidence)")
```

## What is Zero-Shot Prompting?

Zero-shot prompting means asking an AI model to perform a task it was never
explicitly trained for, without giving it any examples first.

You simply describe what you want and the model uses its general understanding
to figure it out.

**Why it matters:** Zero-shot is the foundation of modern LLM usage.
When you write a prompt to ChatGPT or any AI without giving examples,
you're doing zero-shot prompting.

**Real-world use case:** Content moderation, email categorisation,
intent detection in chatbots.
