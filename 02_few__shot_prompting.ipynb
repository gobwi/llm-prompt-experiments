```python
# FEW-SHOT PROMPTING
# We give the model 2-3 examples (shots) before asking our real question.
# This teaches the model the pattern we want it to follow.

from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
print("Model loaded!")
```

```python
# Few-shot example: Sentiment classification
# We show the model examples of the format we want, then ask it to follow that pattern

few_shot_prompt = """Review: "This movie was absolutely fantastic!"
Sentiment: Positive

Review: "Terrible experience, would not recommend."
Sentiment: Negative

Review: "It was okay, nothing special."
Sentiment: Neutral

Review: "I loved every single moment of this film!"
Sentiment:"""

result = generator(few_shot_prompt, max_new_tokens=5, num_return_sequences=1)
print("Prompt given to model:")
print(few_shot_prompt)
print("\nModel output:")
print(result[0]['generated_text'])
```

```python
# Another few-shot example: Text formatting
# Teaching the model to extract structured info from unstructured text

extraction_prompt = """Extract the name and city from the text.

Text: "John Smith lives in Mumbai and works as an engineer."
Name: John Smith, City: Mumbai

Text: "Priya Sharma recently moved to Bangalore from Delhi."
Name: Priya Sharma, City: Bangalore

Text: "Rahul Gupta is a student at a university in Chennai."
Name:"""

result = generator(extraction_prompt, max_new_tokens=20, num_return_sequences=1)
print(result[0]['generated_text'])
```

## What is Few-Shot Prompting?

Few-shot prompting means giving the model 2–5 examples of the exact
format or pattern you want before asking your real question.

Unlike zero-shot (no examples), few-shot "teaches" the model your
specific format through demonstration.

**Why it matters:** Few-shot dramatically improves accuracy for
structured tasks like data extraction, classification, and formatting.

**Real-world use case:** Extracting structured data from documents,
building consistent AI response formats, automating data pipelines.
